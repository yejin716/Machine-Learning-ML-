```python
#결측치 대체 : 평균값 대체 
#SimpleImputer 

from sklearn.impute import SimpleImputer 

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
x_train[['horsepower']] = imputer.fit_transform(x_train[['horsepower']])
```

```python
x_train = x_train.iloc[:, 1:]
x_test = x_test.iloc[:, 1:]

```

```python
#데이터 모델링(모형 구축) >> train(tr/ val)

from sklearn.model_selection import train_test_split

x_tr, x_val, y_tr, y_val=train_test_split(x_train, y_train, test_size=0.3)
```

```python
print(x_tr.shape)
print(x_val.shape)

print(y_tr.shape)
print(y_val.shape)
```

```python
#정규화(scaling)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_tr)

x_tr = scaler.transform(x_tr)
x_val = scaler.transform(x_val)
x_test = scaler.transform(x_test)
```

```python
#모델링 KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean') #유클리드
knn.fit(x_tr, y_tr.values.ravel())
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=10)
dt.fit(x_tr, y_tr)
```

```python
#모델 평가
pred_val = knn.predict(x_val)
```

```python
pred_val
```

```python
pred_val_proba_knn = knn.predict_proba(x_val)
pred_val_proba_dt = dt.predict_proba(x_val)
```

```python
pred_val_proba_knn[:, 1]
```

```python
#roc_auc 
from sklearn.metrics import roc_auc_score

score_knn = roc_auc_score(y_val, pred_val_proba_knn[:, 1])
score_dt = roc_auc_score(y_val, pred_val_proba_dt[:, 1])

print(score_dt, score_knn)
```

```python
best_model = None
best_score = 0 

for i in range(2, 10):
    model = KNeighborsClassifier(n_neighbors=i, metric='euclidean')
    model.fit(x_tr, y_tr.values.ravel())
    pred_val_proba = model.predict_proba(x_val)
    score = roc_auc_score(y_val, pred_val_proba[: ,1])
    print(i, '개의 이웃 확인:', score)
    if best_score <= score:
        best_model = model
        best_score = score
```
