```python
def outlier(df, col):
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    # print(IQR)
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    upper_condition = df[col] > upper_bound
    lower_condition = df[col] < lower_bound

    print('outlier : %.2f   %.2f ' % (upper_bound, lower_bound))
    print('outlier : %.0f' %len(df[upper_condition | lower_condition]))
    print('outlier : %.2f' %(len(df[upper_condition | lower_condition]) / len(df)))
```

```python
outlier(df, 'MEDV')
```

```python
#    
def get_outlier_prop(df, col):
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    upper_condition = df[col] > upper_bound
    lower_condition = df[col] < lower_bound

    outlier_count = len(df[upper_condition | lower_condition])
    total_count = len(df)
    outlier_prop = outlier_count / total_count

    print('outlier : %.2f ' %(outlier_prop))
```

```python
get_outlier_prop(df, 'MEDV')
```

```python
#     ()
def get_outlier_prop_v3(df):
    Q1, Q3 = df.quantile([0.25, 0.75])
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    condition = (df > upper_bound) | (df < lower_bound)
    outliers = df[condition]
    return str(round(100 * len(outliers) / len(df),2)) + '%'
```

```python
df.apply(get_outlier_prop_v3)
```

```python
def outlier_remove(df, col):
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    condition = (df[col] <= upper_bound) & (df[col] >= lower_bound)
    data = df[condition]

    return data

outlier_remove(df, 'MEDV')
```
