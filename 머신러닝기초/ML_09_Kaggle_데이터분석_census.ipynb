```python
#범주형 데이터를 라벨인코딩
from sklearn.preprocessing import LabelEncoder

x = pd.concat([x_train, x_test])

for col in col_cat:
    le = LabelEncoder()
    le.fit(x[col])
    x_train[col] = le.transform(x_train[col])
    x_test[col] = le.transform(x_test[col])
```

```python
#모델링 (모델 구축)

from sklearn.model_selection import train_test_split

x_tr, x_val, y_tr, y_val = \
train_test_split(x_train, y_train, test_size=0.3)
```

```python
#데이터 정규화 - 데이터 스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_tr[col_num] = scaler.fit_transform(x_tr[col_num])
x_val[col_num] = scaler.transform(x_val[col_num])
x_test[col_num] = scaler.transform(x_test[col_num])
```

```python
y_trv = y_tr.values.ravel() #평탄화
```

```python
#선형 모델 사용하지 않음 (LogisticRegressor 사용 안할 결심)
#라벨인코딩을 했기 때문

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_tr,y_trv)
```

1. n_estimators : 결정 트리 개수(default=100)
2. max_depth : 트리 깊이 최대값 (default=None)
3. max_features : 최적 분리 위해 고려해야할 변수 개수(default=auto)
-----------------------------------------------------
4. min_samples_split : node분리에 필요한 최소 샘플 수 (default=2)
5. min_samples_leaf : leaf 노드에 필요한 최소 샘플 수 (default=1)

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(x_tr, y_trv)
```

1. n_estimators : boosting할 결정트리 개수(default=100)
2. max_depth : 트리 깊이 최대값 (default=None)
3. learning rate : 학습률 (default=0.1)
4. lambda : L2 규제 (default=1)
5. alpha : L1 규제 (default=0)
----------------------------------------------
7. min_chid_weight : 자식노드에 필요한 가중치 합의 최소값 (default=1)
8. subsample: 각 트리에 사용할 데이터 샘플 비율 (0.5~1) (default=1)
9. colsample_bytree: 각 트리 생성시, 사용되는 feature(column 열) 비율 의미

```python
xgb_tune = XGBClassifier(n_estimators=1000, learning_rate = 0.1, max_depth=10)
xgb_tune.fit(x_tr, y_trv, early_stopping_rounds=50,
             eval_metric='auc', eval_set=[(x_val, y_val)], verbose=10)

#early_stopping_rounds=50 조기종료 (50회까지 돌려 성능 향상 없으면 훈련중지)
#eval_metrics = 'auc' 평가(evaluation) 지표 auc(1에 가까울 수록 분류 정확도가 높음)
#eval_set=[(x_val, y_val)] >> 검증셋
#verbose=10 훈련과정 출력 옵션
```

```python
#ROC_AUC

from sklearn.metrics import roc_auc_score

rf.predict(x_val)
```

```python
pred_rf = rf.predict_proba(x_val)
pred_xbg = xgb.predict_proba(x_val)
pred_xbg_tune = xgb_tune.predict_proba(x_val)
```

```python
pred_rf[:]
# 앞은 y_val, 뒤 pred_rf값
```

```python
score_rf = roc_auc_score(y_val, pred_rf[:, 1])
score_xgb = roc_auc_score(y_val, pred_xbg[:, 1])
score_xgb_tune = roc_auc_score(y_val, pred_xbg_tune[:, 1])
```
