```python
#  
from sklearn.preprocessing import LabelEncoder

x = pd.concat([x_train, x_test])

for col in col_cat:
    le = LabelEncoder()
    le.fit(x[col])
    x_train[col] = le.transform(x_train[col])
    x_test[col] = le.transform(x_test[col])
```

```python
# ( )

from sklearn.model_selection import train_test_split

x_tr, x_val, y_tr, y_val = \
train_test_split(x_train, y_train, test_size=0.3)
```

```python
#  -  
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_tr[col_num] = scaler.fit_transform(x_tr[col_num])
x_val[col_num] = scaler.transform(x_val[col_num])
x_test[col_num] = scaler.transform(x_test[col_num])
```

```python
y_trv = y_tr.values.ravel() #
```

```python
#    (LogisticRegressor   )
#  

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_tr,y_trv)
```

1. n_estimators :   (default=100)
2. max_depth :    (default=None)
3. max_features :      (default=auto)

4. min_samples_split : node     (default=2)

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(x_tr, y_trv)
```

1. n_estimators : boosting  (default=100)
2. max_depth :    (default=None)
3. learning rate :  (default=0.1)
4. lambda : L2  (default=1)
5. alpha : L1  (default=0)
----------------------------------------------
7. min_chid_weight :      (default=1)
8. subsample:       (0.5~1) (default=1)

```python
xgb_tune = XGBClassifier(n_estimators=1000, learning_rate = 0.1, max_depth=10)
xgb_tune.fit(x_tr, y_trv, early_stopping_rounds=50,
             eval_metric='auc', eval_set=[(x_val, y_val)], verbose=10)

#early_stopping_rounds=50  (50     )
#eval_metrics = 'auc' (evaluation)  auc(1     )
#eval_set=[(x_val, y_val)] >> 
#verbose=10   
```

```python
#ROC_AUC

from sklearn.metrics import roc_auc_score

rf.predict(x_val)
```

```python
pred_rf = rf.predict_proba(x_val)
pred_xbg = xgb.predict_proba(x_val)
pred_xbg_tune = xgb_tune.predict_proba(x_val)
```

```python
pred_rf[:]
#  y_val,  pred_rf
```

```python
score_rf = roc_auc_score(y_val, pred_rf[:, 1])
score_xgb = roc_auc_score(y_val, pred_xbg[:, 1])
score_xgb_tune = roc_auc_score(y_val, pred_xbg_tune[:, 1])
```
