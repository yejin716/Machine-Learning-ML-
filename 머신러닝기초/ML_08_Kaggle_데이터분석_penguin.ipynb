```python
# one-hot encoding (문자 >> 숫자)
# x_train 과 x_test 모두 인코딩 해줘야함

x = pd.concat([x_train, x_test])
x
```

```python
x.columns
```

```python
#범주형 데이터
cols = ['species', 'island', 'sex']

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(x[cols])

x_train_ohe = ohe.transform(x_train[cols])
x_test_ohe = ohe.transform(x_test[cols])
```

```python
print(x_train_ohe)
#sparse matrix 희소 행렬
```

```python
x_train_ohe.todense()
#todense() : 희소행렬 >> 밀집 행렬(dense matrix) 변환
```

```python
ohe.get_feature_names_out()
```

```python
pd.DataFrame(x_train_ohe.todense(), columns=ohe.get_feature_names_out())
```

```python
x_train_ohe = pd.DataFrame(x_train_ohe.todense(), columns=ohe.get_feature_names_out())
x_test_ohe = pd.DataFrame(x_test_ohe.todense(), columns=ohe.get_feature_names_out())
```

```python
x_train_ohe.head(3)
```

```python
x_test_ohe.head(3)
```

```python
x_train.columns
```

```python
col_num = ['bill_length_mm', 'bill_depth_mm','flipper_length_mm']
```

```python
x_train_fin = pd.concat([x_train[col_num], x_train_ohe], axis=1)
x_test_fin = pd.concat([x_test[col_num], x_test_ohe], axis=1)
```

```python
x_train_fin.head(3)
```

```python
#데이터 모형 구축
from sklearn.model_selection import train_test_split

x_tr, x_val, y_tr, y_val = train_test_split(x_train_fin, y_train, test_size=0.3)
```

```python
x_tr
```

```python
x_tr[col_num]
```

```python
x_val[col_num]
```

```python
x_test_fin[col_num]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x_tr[col_num])
```

```python
x_tr[col_num] = scaler.transform(x_tr[col_num])
x_val[col_num] = scaler.transform(x_val[col_num])
x_test[col_num] = scaler.transform(x_test[col_num])
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_tr, y_tr)
```

```python
pred_val = lr.predict(x_val)
pred_val
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
x_train_fin.columns
```

```python
pd.DataFrame(lr.coef_, index=x_train_fin.columns, columns=['coef'])
```

```python
coef = pd.DataFrame(lr.coef_, index=x_train_fin.columns, columns=['coef'])
```

```python
coef.rename_axis('feature', axis=0)
# coef.index.name = 'feature'
```

```python
#데이터 모형 평가
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_val, pred_val)
rmse = mean_squared_error(y_val, pred_val, squared=False)

print(f'MSE : {mse:.3f} RMSE : {rmse : .3f}')
```
